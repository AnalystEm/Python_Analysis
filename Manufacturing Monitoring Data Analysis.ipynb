```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

df = pd.read_excel('Manufacturing Rig Data.xlsx')

df.head()
```

```python
df.shape
```

```python
df.info()
```

# Data Cleaning

```python
# Find the number of null values for all columns
df.isnull().sum()
```

```python
# Find the number of duplicate data
df.duplicated().sum()
```

# Summary Statistics

```python
df.describe()
```

```python
mean_values = df.mean()
median_values = df.median()

summary_table = pd.DataFrame({'Mean': mean_values, 'Median': median_values})

print(summary_table)
```

```python
std_dev = df.std()

summary_table = pd.DataFrame({'Standard Deviation': std_dev})

print(summary_table)
```

```python
from scipy.stats import skew

skewness_values = df.skew()

summary_table = pd.DataFrame({'Skewness': skewness_values})

print(summary_table)
```

# Outliers

```python
data1 = df['Motor power W']
data2 = df['Efficiency factor']
data3 = df['Pressure bar 1']
data4 = df['Pressure bar 2']

fig, axes = plt.subplots(1, 4, figsize=(12, 4))

axes[0].boxplot(data1)
axes[0].set_title('Motor power W')

axes[1].boxplot(data2)
axes[1].set_title('Efficiency factor')

axes[2].boxplot(data3)
axes[2].set_title('Pressure bar 1')

axes[3].boxplot(data4)
axes[3].set_title('Pressure bar 2')

plt.tight_layout()

plt.show()
```

# Correlation Analysis

```python
corr_matrix = df.corr()

fig, ax = plt.subplots(figsize=(12, 5))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.regplot(x='Cooling efficiency', y='Cooling power', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'black'}, ax=axes[0])
axes[0].set_title('Cooling efficiency vs Cooling power')

sns.regplot(x='Cooling efficiency', y='Temperature 1', data=df, scatter_kws={'color':'red'}, line_kws={'color':'black'}, ax=axes[1])
axes[1].set_title('Cooling efficiency vs Temperature')

plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.regplot(x='Volume flow l/min 2', y='Pressure bar 5', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'black'}, ax=axes[0])
axes[0].set_title('Volume flow vs Pressure bar')

sns.regplot(x='Cooling efficiency', y='Volume flow l/min 2', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'black'}, ax=axes[1])
axes[1].set_title('Cooling efficiency vs Volume flow')

plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.regplot(x='Cooling power', y='Motor power W', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'black'}, ax=axes[0])
axes[0].set_title('Cooling power vs Motor power')

sns.regplot(x='Motor power W', y='Temperature 1', data=df, scatter_kws={'color':'red'}, line_kws={'color':'black'}, ax=axes[1])
axes[1].set_title('Motor power vs Temperature')

plt.tight_layout()

plt.show()
```

# Time Series Analysis

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Cooling efficiency', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:green')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    

plot_df(df, x=df['Date'], y=df['Cooling efficiency'], title='Cooling Efficiency Trend Over Time')
```
